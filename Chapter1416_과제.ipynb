[문제 2: 붕어빵 주문 받기]

손님이 주문한 붕어빵의 종류와 개수를 입력받는 기능을 구현하세요.

요구사항:
- 손님이 주문할 붕어빵의 종류와 개수를 입력받습니다.
- 주문 내용을 딕셔너리에 저장하세요.
- 주문 내역을 출력해주세요.

```python
order = {}

print("붕어빵 주문을 시작합니다. '종료'를 입력하면 주문이 완료됩니다.")

while True:
    what = input("붕어빵 종류를 입력하세요: ")
    if what.lower() == '종료':
        break
    
    many = input(f"{what}의 개수를 입력하세요: ")
    if many.isdigit():
        many = int(many)
        if what in order:
            order[what] += many
        else:
            order[what] = many
    else:
        print("유효한 숫자를 입력해 주세요.")

print("\n주문 내역:")
for item, many in order.items():
    print(f"{item} : {many}개")
```

[문제 3: 붕어빵 주문 처리]

손님의 주문을 처리하여 붕어빵 재고를 업데이트하는 기능을 구현하세요.

요구사항:
- 손님의 주문 내용을 기반으로 재고를 업데이트합니다.
- 재고가 부족할 경우, 손님에게 재고 부족을 알리고 재고를 감소시키지 않습니다.
- 재고가 충분할 경우, 주문한 만큼 재고를 감소시키고 판매를 완료합니다.
- 판매가 완료된 경우 판매된 붕어빵 맛과 개수를 출력하세요.

```python
inventory = {
    "팥붕어빵": 10,
    "슈크림붕어빵": 8,
    "초코붕어빵": 5
}

def process_order(order):
    for kind, quantity in order.items():
        if kind in inventory:
            if inventory[kind] >= quantity:
                inventory[kind] -= quantity
                print(f"{kind} : {quantity}개 판매 완료되었습니다.")
            else:
                print(f"{kind}의 재고가 부족합니다. 현재 재고: {inventory[kind]}개")
        else:
            print(f"{kind}은(는) 재고 목록에 없습니다.")
    
    print("\n현재 재고 상태:")
    for item, qty in inventory.items():
        print(f"{item} : {qty}개")

order = {
    "팥붕어빵": 2,
    "슈크림붕어빵": 10, 
    "초코붕어빵": 3
}

process_order(order)
```

[문제 4: 여러 손님 주문 처리]

여러 손님들의 주문을 처리할 수 있도록 시스템을 확장하세요.

요구사항:
- 손님이 "종료"를 입력할 때까지 주문을 계속 받습니다.
- 각 주문마다 재고를 업데이트합니다.
- 현재 재고를 화면에 출력합니다.

```python
inventory = {
    "팥붕어빵": 10,
    "슈크림붕어빵": 8,
    "초코붕어빵": 5
}

def display_inventory():
    print("\n현재 재고 상태:")
    for item, qty in inventory.items():
        print(f"{item} : {qty}개")
    print()  

def process_order(order):
    for kind, quantity in order.items():
        if kind in inventory:
            if inventory[kind] >= quantity:
                inventory[kind] -= quantity
                print(f"{kind} : {quantity}개 판매 완료되었습니다.")
            else:
                print(f"{kind}의 재고가 부족합니다. 현재 재고: {inventory[kind]}개")
        else:
            print(f"{kind}은(는) 재고 목록에 없습니다.")

def take_order():
    print("붕어빵 주문을 시작합니다. '종료'를 입력하면 주문이 완료됩니다.")

    while True:
        order = {}
        while True:
            kind = input("붕어빵 종류를 입력하세요 (종료: '종료'): ")
            if kind.lower() == '종료':
                return 
            
            if kind in inventory:
                quantity = input(f"{kind}의 개수를 입력하세요: ")
                if quantity.isdigit():
                    quantity = int(quantity)
                    if quantity > 0:
                        if kind in order:
                            order[kind] += quantity
                        else:
                            order[kind] = quantity
                    else:
                        print("개수는 1 이상의 정수여야 합니다.")
                else:
                    print("유효한 숫자를 입력해 주세요.")
            else:
                print(f"{kind}은(는) 재고 목록에 없습니다.")
        
        process_order(order) 
        
        display_inventory()  

take_order()
```

[문제 5: 관리자 모드]

관리자 모드를 통해 붕어빵 재고를 추가할 수 있는 기능을 구현하세요.

요구사항:
- 관리자는 특정 붕어빵의 재고를 추가할 수 있습니다.
- "종료"를 입력하면 관리자 모드를 종료합니다.
- 추가된 재고를 업데이트하고 현재 재고를 화면에 출력합니다.

```python
# 초기 붕어빵 재고
inventory = {
    "팥붕어빵": 10,
    "슈크림붕어빵": 8,
    "초코붕어빵": 5
}

def display_inventory():
    print("\n현재 재고 상태:")
    for item, qty in inventory.items():
        print(f"{item} : {qty}개")
    print()  

def process_order(order):
    for kind, quantity in order.items():
        if kind in inventory:
            if inventory[kind] >= quantity:
                inventory[kind] -= quantity
                print(f"{kind} : {quantity}개 판매 완료되었습니다.")
            else:
                print(f"{kind}의 재고가 부족합니다. 현재 재고: {inventory[kind]}개")
        else:
            print(f"{kind}은(는) 재고 목록에 없습니다.")
    
    display_inventory()  # 현재 재고 상태를 출력

def take_order():
    print("붕어빵 주문을 시작합니다. '종료'를 입력하면 주문이 완료됩니다.")

    while True:
        order = {}
        while True:
            kind = input("붕어빵 종류를 입력하세요 (종료: '종료'): ")
            if kind.lower() == '종료':
                return  
            
            if kind in inventory:
                quantity = input(f"{kind}의 개수를 입력하세요: ")
                if quantity.isdigit():
                    quantity = int(quantity)
                    if quantity > 0:
                        if kind in order:
                            order[kind] += quantity
                        else:
                            order[kind] = quantity
                    else:
                        print("개수는 1 이상의 정수여야 합니다.")
                else:
                    print("유효한 숫자를 입력해 주세요.")
            else:
                print(f"{kind}은(는) 재고 목록에 없습니다.")
        
        process_order(order)  

def admin_mode():
    print("관리자 모드에 진입했습니다. '종료'를 입력하면 관리자 모드를 종료합니다.")
    
    while True:
        kind = input("재고를 추가할 붕어빵 종류를 입력하세요 (종료: '종료'): ")
        if kind.lower() == '종료':
            break
        
        if kind in inventory:
            quantity = input(f"{kind}의 추가할 개수를 입력하세요: ")
            if quantity.isdigit():
                quantity = int(quantity)
                if quantity > 0:
                    inventory[kind] += quantity
                    print(f"{kind}의 재고가 {quantity}개 추가되었습니다.")
                else:
                    print("추가할 개수는 1 이상의 정수여야 합니다.")
            else:
                print("유효한 숫자를 입력해 주세요.")
        else:
            print(f"{kind}은(는) 재고 목록에 없습니다.")
        
        display_inventory() 

def main():
    while True:
        print("1. 손님 주문하기")
        print("2. 관리자 모드")
        print("3. 종료")
        choice = input("선택하세요: ")
        
        if choice == '1':
            take_order()  
        elif choice == '2':
            admin_mode()  
        elif choice == '3':
            print("프로그램을 종료합니다.")
            break
        else:
            print("유효하지 않은 선택입니다. 다시 시도해 주세요.")

main()
```

[문제 6: 매출 확인]

붕어빵 판매 가격을 설정하고, 판매한 붕어빵의 매출을 계산하세요.

요구사항:
- 각 붕어빵의 가격을 딕셔너리로 관리합니다.
- 판매된 붕어빵의 종류와 개수를 딕셔너리로 관리합니다.
- 총 매출을 계산하여 출력합니다.

초기 설정:
- 가격
    - "팥붕어빵": 1000
    - "슈크림붕어빵": 1200
    - "초코붕어빵": 1500

- 판매 개수
    - "팥붕어빵": 0
    - "슈크림붕어빵": 0
    - "초코붕어빵": 0

```python
price_dict = {
    "팥붕어빵": 1000,
    "슈크림붕어빵": 1200,
    "초코붕어빵": 1500
}

sold_dict = {
    "팥붕어빵": 0,
    "슈크림붕어빵": 0,
    "초코붕어빵": 0
}

sold_dict["팥붕어빵"] = 5  
sold_dict["슈크림붕어빵"] = 3  
sold_dict["초코붕어빵"] = 2  

def calculate_revenue(price_dict, sold_dict):
    total_revenue = 0
    for bread_type, count in sold_dict.items():
        total_revenue += price_dict[bread_type] * count
    return total_revenue

total_revenue = calculate_revenue(price_dict, sold_dict)
print(f"총 매출: {total_revenue}원")
```
